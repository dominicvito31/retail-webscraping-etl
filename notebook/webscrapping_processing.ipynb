# Import Libraries

```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
import time
```

# Extract

## Web Scraping

```python
# Variabel untuk menyimpan data
nama = []
harga = []
diskon = []
jumlah_terjual = []
kota = []

for page in range (1, 4):
    # Buat driver
    driver = webdriver.Chrome()

    # Akses website
    url = f'https://www.lazada.co.id/catalog/?from=hp_categories&page={page}&q=Peralatan%20Dapur%20Kecil&service=all_channel&spm=a211g0.flashsale.cate_8.2.3ccf6nNb6nNbxr&src=all_channel'
    driver.get(url)

    # Tunggu website selesai load
    time.sleep(5)

    # Fetch and clean HTML
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    # Mangambil semua data dalam boxes
    boxes = soup.find_all('div', {'class': 'buTCk'})
    # Mengambil data berdasarkan kategori
    for box in boxes:
        # Kategori nama
        try:
            box_nama = box.find('div', {'class': 'RfADt'})
            nama.append(box_nama.get_text())
        except:
            nama.append(None)
        # Kategori harga
        try:
            box_harga = box.find('span', {'class': 'ooOxS'})
            harga.append(box_harga.get_text())
        except:
            harga.append(None)
        # Kategori diskon
        try:
            box_diskon = box.find('span', {'class': 'IcOsH'})
            diskon.append(box_diskon.get_text())
        except:
            diskon.append(None)
        # Kategori jumlah terjual
        try:
            box_terjual = box.find('span', {'class': '_1cEkb'})
            jumlah_terjual.append(box_terjual.get_text())
        except:
            jumlah_terjual.append(None)
        # Kategori kota asal
        try:
            box_kota = box.find('span', {'class': 'oa6ri'})
            kota.append(box_kota.get_text())
        except:
            kota.append(None)

    # Matikan webdriver
    driver.quit()
```

## Create DataFrame

```python
# Buat DataFrame kosong
df = pd.DataFrame()

# Masukkan lists ke dalam DataFrame
df['nama_produk'] = nama
df['harga_produk'] = harga
df['diskon'] = diskon
df['banyaknya_terjual'] = jumlah_terjual
df['kota_toko'] = kota

# Cek DataFrame
df
```

## Export Raw Data

```python
df.to_csv('data/scraped_products.csv', index=False)
```

### Clean Data

#### Pengubahan data pada kolom harga_produk

```python
# Hapus karakter yang bukan angka
df['harga_produk'] = df['harga_produk'].str.replace('Rp', '')
df['harga_produk'] = df['harga_produk'].str.replace(',', '')
df['harga_produk'] = df['harga_produk'].str.replace('.', '')

# Mengubah tipe data ke numerik
df['harga_produk'] = df['harga_produk'].astype(int)

# Cek tipe data harga_produk
df['harga_produk'].dtypes
```

#### Pengubahan data pada kolom diskon

```python
# Hapus karakter yang bukan angka
df['diskon'] = df['diskon'].str.replace('% Off', '')

# Mengubah NaN menjadi '0'
df['diskon'] = df['diskon'].fillna('0')

# Mengubah tipe data ke numerik
df['diskon'] = df['diskon'].astype(int)

# Cek tipe data diskon
df['diskon'].dtypes
```

#### Pengubahan data pada kolom banyaknya_terjual

```python
# Hapus karakter yang bukan angka
df['banyaknya_terjual'] = df['banyaknya_terjual'].str.replace(' sold', '')
df['banyaknya_terjual'] = df['banyaknya_terjual'].str.replace('.', '')
df['banyaknya_terjual'] = df['banyaknya_terjual'].str.replace('K', '00')

# Mengubah NaN menjadi '0'
df['banyaknya_terjual'] = df['banyaknya_terjual'].fillna('0')

# Mengubah tipe data ke numerik
df['banyaknya_terjual'] = df['banyaknya_terjual'].astype(int)

# Cek tipe data banyaknya_terjual
df['banyaknya_terjual'].dtypes
```
